```python
import time
start=time.time()
f_per_air=[]
for index in data.index:
    res=df_per_air.loc[index].value_counts().index
    #if len(res)>2:
    #    air=res[1]
    #else:
    #    air='0'
    air=res[0]
    f_per_air.append(air)
end=time.time()
end-start
```
